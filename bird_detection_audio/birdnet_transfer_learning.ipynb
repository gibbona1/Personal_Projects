```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
birdnet_output_dir = '../../../birdnet_output'
birdnet_txt_files = os.listdir(birdnet_output_dir)
```

```python
birdnet_txt_files[0]
```

```python
full_df = pd.DataFrame()
for f in birdnet_txt_files:
    row_df = pd.read_table(os.path.join(birdnet_output_dir, f))
    row_df['file_name'] = f.split('.')[0]+'.wav'
    full_df = pd.concat([full_df, row_df])
```

```python
full_df
```

```python
lab_df = pd.read_csv('../../neal_data/labels.csv')
lab_df
```

```python
merge_df = pd.merge(full_df, lab_df, on = 'file_name')
merge_df.drop_duplicates(subset = ['file_name', 'Common Name'], inplace = True)
merge_df.drop_duplicates(subset = ['file_name', 'class_label'], inplace = True)

def replace_start(x):
    return x.replace("Common ", "").replace("Eurasian ", "").replace("European ", "").replace("Northern ", "").replace("Greater ", "").replace("Great ", "")
merge_df['Common Name'] = merge_df['Common Name'].apply(replace_start)
merge_df['class_label'] = merge_df['class_label'].apply(replace_start)
merge_df
```

```python
np.mean(merge_df.groupby('file_name').apply(lambda x: (x['Common Name'] == x['class_label']).any()))
```

```python
merge_df.groupby('file_name').apply(lambda x: x[(x['Common Name'] != x['class_label'])])
```
