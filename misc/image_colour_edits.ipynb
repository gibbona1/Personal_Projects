```python
# Open the webp image
image = Image.open(im_path)

# Convert the image to RGBA (if not already in that mode)
image = image.convert('RGBA')

# Get the data of the image
data = image.getdata()

# Create a new list to hold the modified pixel data
new_data = []

# Iterate through each pixel
for item in data:
    # Change all red-ish or skin colour pixels to white
    #if item[0] > 50 and item[0] > item[1]+10 and item[0] > item[2]+10:
        #make transparent
    #    new_data.append((255, 255, 255, 0))
        #new_data.append((255, 255, 255, item[3]))
    #else if item is white
    #elif item[:3] != (0, 0, 0):
    #    new_data.append((255, 255, 255, 0))  # Change to white
    if item[2] > 50 and item[2] > item[0] + 10 and item[2] > item[1]+10:
        new_data.append((255, 255, 255, 0))
    elif any([item[0] > 80, item[1] > 80, item[2] > 80]):
        new_data.append((0, 0, 0, 255))
    elif item == (0, 0, 0, 255):
        new_data.append(item)
    elif item[3] == 0:
        new_data.append(item)
    elif all([item[0] < 50, item[1] < 50, item[2] < 50]):
        new_data.append((0, 0, 0, 255))
    elif all([50< item[0] <100, 50< item[1] <100, 50 < item[2] < 100]):
        new_data.append((255, 255, 255, 0))
    #elif item[:3] != (0, 0, 0) and item[:3] != (255, 255, 255):
    #    new_data.append((255, 255, 255, item[3]))  # Change to white
    else:
        new_data.append((item[0], item[1], item[2], 255))  # Keep the original pixel

# Update the image data with the new data
image.putdata(new_data)

# Save the modified image as a PNG file
image.save('output_image.png')
```
